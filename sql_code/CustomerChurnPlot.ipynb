```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

columnsdelete=['total_spent_january',
 'total_spent_february',
 'total_spent_march',
 'total_spent_april',
 'total_spent_may',
 'total_spent_june',
 'total_spent_july',
 'total_spent_august',
 'total_spent_september',
 'overall_total_spent']

df=pd.read_csv(r'..\csv_files\CustomerChurnTable.csv')
df=df.sort_values(by='overall_total_spent',ascending=False).head(20)
df=df.drop(labels=columnsdelete,axis=1)
df.columns.to_list()
```

```python
df_long = pd.melt(df, id_vars=['top_customerid'], 
                   var_name='month', 
                   value_name='total_spent')

df_long['month'] = df_long['month'].str.replace('total_spent_', '', regex=False)

plt.figure(figsize=(14, 8))
sns.barplot(data=df_long, x='top_customerid', y='total_spent', hue='month', palette='Paired')

plt.title('Monthly Spending by Customer')
plt.xlabel('Customer_id')
plt.ylabel('Total Spent')
plt.xticks(rotation=45)
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```
